```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://movies.yahoo.com.tw/movie_intheaters.html'
r = requests.get(url)
r.encoding = 'utf-8'
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
print(soup.prettify())
```

```python
print(soup.find(class_="release_box").p.text)
```

```python
max_page = 5
for page_number in range(1, max_page+1, 1):
    url = "https://movies.yahoo.com.tw/movie_intheaters.html"
    payload = {'page':str(page_number)}
    resp = requests.get(url, params = payload)
    resp.encoding = 'utf-8'
    soup = BeautifulSoup(resp.text, 'lxml')
    movie_list = soup.find('ul', class_='release_list').find_all('li')
    for p in movie_list:
        print("-----------------------------------------------------------------")
        movie_name = p.find('div', class_='release_movie_name')
        print("名稱：",movie_name.a.text)
        
        level_box = p.find(class_='leveltext')
        print('期待度：',level_box.span.string)
        
        satisfaction = p.find('span', class_='count')
        print("滿意度：", satisfaction['data-num'])
        
        movie_info = p.find('div', class_='release_text')
        print(movie_info.span.string.replace('\t',''))
```

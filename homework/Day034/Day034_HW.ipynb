```python
import requests
from bs4 import BeautifulSoup
```

```python
def proxy_ips_fun(url):
    res = requests.get(url, 'html.parser')
    soup = BeautifulSoup(res.text)
    tbody = soup.find('tbody')
    tr = tbody.find_all('tr')
    proxy_ips = []
    for x in tr:
        for i, y in enumerate(x):
            if i == 0:
                proxy_ips.append(y.text)
    return proxy_ips
    
```

```python
proxy_ips_fun('https://free-proxy-list.net/')
```
